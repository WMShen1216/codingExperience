```python
import itertools
import random
import pandas as pd
import csv
```

```python
def form_wards(blocks, start_floor, end_floor):
    blocklist = blocks
    floorlist = [floor for floor in range(start_floor,end_floor+1)]

    unique_combinations = []

    for block in range(len(blocklist)):
        for floor in range(len(floorlist)):
            unique_combinations.append(blocklist[block]+str(floorlist[floor]))

    return unique_combinations
```

```python
AB_2_6 = form_wards (["A", "B", "C","D"],2,6)
AB_7_10 = form_wards (["A", "B", "C"],7,10)
AB_others = ['A11','C11','ALG','A1','M5']
EF_2_10 = form_wards (["E", "F", "G","H"],2,10)
EF_others = ['E11','G11','G1','H1']
R_1_10 = form_wards (["R"],1,10)

# create a list that contains all the wards except H1
ward_list = list(itertools.chain(AB_2_6,AB_7_10,AB_others,EF_2_10,EF_others,R_1_10))
ward_list.remove('H1')
ward_list.sort()
```

```python
# simulate requests
def simu_requ(n, ward_list):
    requests = random.choices(ward_list, k = n)
    return requests
```

```python
requests = simu_requ(100, ward_list)
print(requests)
```

```python
requ_df = pd.DataFrame(columns = ['requests'])
```

```python
# sort the requests by wards
for i in range(len(requests)):
    if requests[i] in list(requ_df.index.values):
        requ_df.loc[requests[i], 'requests'] += 1
    else:
        requ_df.loc[requests[i], 'requests'] = 1
```

```python
service_time = 2 # 2 minutes per requests
requ_df['time'] = requ_df['requests'] * service_time
```
